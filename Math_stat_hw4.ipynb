Подключим нужные библиотеки для моделирования и визуализации расчетов

```python
import numpy as np
import matplotlib.pyplot as plt
```

Генерируем случайные выборки из равномерного распределения [0, 10]

```python
theta_0 = 10 
sample_sizes = [10, 50, 100, 500, 1000, 5000]
num_experiments = 1000 
```

Эксперемент 

```python
def estimate_theta(sample):
    return np.max(sample)

# Проведение экспериментов
estimates = {n: [] for n in sample_sizes}

for n in sample_sizes:
    for _ in range(num_experiments):
        sample = np.random.uniform(0, theta_0, n)
        theta_hat = estimate_theta(sample)
        estimates[n].append(theta_hat)
```

Графики

```python
plt.figure(figsize=(12, 8))

for n in sample_sizes:
    plt.hist(estimates[n], bins=30, alpha=0.5, label=f'n={n}', density=True)

plt.axvline(x=theta_0, color='r', linestyle='--', label=f'θ_0 = {theta_0}')
plt.xlim(9, 11)  # Устанавливаем диапазон по оси x
plt.xlabel('Оценка θ')
plt.ylabel('Плотность')
plt.title('Распределение оценок θ для различных размеров выборки')
plt.legend()
plt.show()

# Проверка асимптотической несмещенности
mean_estimates = [np.mean(estimates[n]) for n in sample_sizes]
plt.figure(figsize=(12, 8))
plt.plot(sample_sizes, mean_estimates, marker='o', linestyle='-', label='Средняя оценка θ')
plt.axhline(y=theta_0, color='r', linestyle='--', label=f'θ_0 = {theta_0}')
plt.xlabel('Размер выборки n')
plt.ylabel('Средняя оценка θ')
plt.title('Средняя оценка θ в зависимости от размера выборки')
plt.legend()
plt.show()
```
